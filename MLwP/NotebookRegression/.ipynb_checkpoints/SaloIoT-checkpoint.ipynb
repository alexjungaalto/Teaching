## 3.3 Linear Regression 

Our goal is to predict the closing price $y^{(i)}$ of Ethereum at some day $i$ based on the Bitcoin closing price $x^{(i)}$ on the very same day. Within linear regression, we try to do this prediction using linear functions which are elements of the hypothesis space

\begin{equation*}
 \mathcal{H} = \{h^{(\mathbf{w})}(\mathbf{x}) = \mathbf{w}^{T} \mathbf{x} \mbox{ for some } \mathbf{w} \in \mathbb{R}\}
\label{eq1}
\tag{1}
\end{equation*}

In particular, we try to predict or approximate $y^{(i)}$ by the function value (predictor) $h^{(\mathbf{w})}(\mathbf{x}^{(i)}) = \mathbf{w}^{T}\mathbf{x}^{(i)}$ using the augmented feature vector $\mathbf{x}^{(i)} = (x^{(i)},1)^{T}$. At first sight, it seems strange to add a constant one to each feature value $x^{(i)}$ (which is the closing price of Bitcoin at day i). However, it will turn out to allow for a larger class of linear functions $h^{(\mathbf{w})}$ which can be used for the prediction. The prediction $h^{(\mathbf{w})}(\mathbf{x}^{(i)})$ will typically incur a non-zero __prediction error__ $y^{(i)} - h^{(\mathbf{w})}(\mathbf{x}^{(i)})$, which we quantify using its square $(y^{(i)} - h^{(\mathbf{w})}(\mathbf{x}^{(i)})^2$. Since we have $N$ data points $(x^{(i)},y^{(i)})$ available, we can compute the average loss which we term the __empirical risk__ 
\begin{equation*}
 \mathcal{E} (\mathbf{w}) = \frac{1}{N}\sum^{N}_{i=1}(y^{(i)} - \mathbf{w}^{T} x^{(i)})^2
\label{eq2}
\tag{2}
\end{equation*}


The optimal predictor $h_{\rm opt}(\cdot)$ is then

\begin{equation*}
h_{\rm opt}(\cdot) = \underset{{h(\mathbf{w})} \in \mathcal{H}}{\operatorname{argmin}} \mathcal{E} (\mathbf{w})
\label{eq3}
\tag{3}
\end{equation*}

Note that every function $h(\mathbf{w}) \in \mathcal{H}$ corresponds to a particular choice of the weight $\mathbf{w}$.

This allows us to rewrite the optimization problem (Eq. \ref{eq3}) as an optimization of the weight vector.

\begin{align}
\mathbf{w}_{\rm opt} &= \underset{ \mathbf{w} \in \ \mathbb{R}^{2}}{\operatorname{argmin}} \{ \frac{1}{N}\sum^{N}_{i=1}(y^{(i)} - \mathbf{w}^{T} \mathbf{x}^{(i)})^2 \}\\
&= \underset{ \mathbf{w} \in \ \mathbb{R}^{2}}{\operatorname{argmin}} \frac{1}{N} {||\mathbf{y}-\mathbf{X}\mathbf{w}||}_2^2,
\label{eq4}
\tag{4}
\end{align}
where
\begin{align}
\mathbf{X} = \begin{bmatrix}\mathbf{x}^{(1)}\\\vdots\\\mathbf{x}^{(n)}\end{bmatrix}\ \text{and }\mathbf{y} = \begin{bmatrix}y^{(1)}\\\vdots\\y^{(n)}\end{bmatrix}\ \mathbf{X},\mathbf{y}\in\mathbb{R}^{n}
\end{align}

The solution of (\ref{eq4}), i.e., the optimal weight which minimizes the empirical risk, is obtained as 

\begin{equation*}
\mathbf{w}_{\rm opt} = (\mathbf{X}^T \mathbf{X})^{-1}\mathbf{X}^T \mathbf{y}
\label{eq5}
\tag{5}
\end{equation*}

However, as a general note, (\ref{eq5}) only works when $\mathbf{X}^T \mathbf{X}$ is invertible.

## 3.4 Polynomial Regression

By looking at the $\bf{Figure\ 3.}$ we can easily see that relation between label (Bitcoin, $\mathbf{y}$) and and feature (Ethereum, $\mathbf{x}$) becomes highly non-linear when the value of each currency increases. In these cases, it's useful to consider a hypothesis space which is constituted by polynomial functions, i.e., 

\begin{equation}
\label{equ_def_poly_hyposapce}
\mathcal{H}^{(d)}_{\rm poly} = \{ h^{(w)}(\cdot): \mathbf{X} \rightarrow \mathbf{Y}: h^{(\mathbf{w})}(\mathbf{x}) = \sum_{r=1}^{d} \mathbf{w}_{r} \mathbf{x}^{r-1} \mbox{, with some } \mathbf{w} =(w_{1},\ldots,w_{d})^{T} \in \mathbb{R}^{d} \}.
\end{equation}  

And just like in task `3.3 Linear Regression` the quality of a predictor is measured by the squared error loss and linear regression amounts to minimizing the  average squared error loss (mean squared error):
\begin{equation} 
\min_{h \in \mathcal{H}_{\rm poly} } \frac{1}{N} \sum_{i=1}^{d} (y^{(i)} - h^{(\mathbf{w})}(\mathbf{x}^{(i)}))^{2} \mbox{, where } h^{(\mathbf{w})}(\mathbf{x})=\sum_{r=1}^{d} \mathbf{w}_{r} \mathbf{x}^{r-1}
\end{equation} 

To better understand how this works, consider a dataset $\mathbf{X}$ consisting of Bitcoin's closing prices and $N$ amount of data rows, represented as a feature vector $x^{(i)}=(x^{(i)})^{T} \in \mathbb{R}^{1}$ where $i=1,\ldots,N$ defines the row position of the data in the given dataset. We are interested in transforming those data points into higher dimensional feature space by using a feature map:

\begin{equation*}
\phi(\mathbf{x}) = (x^d, x^{d-1},..,x^1, x^0)
\label{eq7}
\tag{7}
\end{equation*}

which takes the original feature $\mathbf{x} \in \mathbb{R}^1$ (i.e. one closing price) as an input and returns a new feature vector $ \mathbf{\hat{x}} \in \mathbb{R}^{d+1} $ with the new dimension of $d+1$. 

In the end, the polynomial feature matrix should look something like this:

\begin{equation*}
    \mathbf{X}_{poly}=\begin{bmatrix}
        (x^{(1)})^{d} & (x^{(1)})^{d-1} & \dots & (x^{(1)})^{1}& (x^{(1)})^{0}\\
        (x^{(2)})^{d} & (x^{(2)})^{d-1} & \dots & (x^{(2)})^{1}& (x^{(2)})^{0}\\
        \vdots &  \vdots & \ddots& \vdots & \vdots\\
        (x^{(N-1)})^{d} & (x^{(N-1)})^{d-1} & \dots & (x^{(N-1)})^{1}& (x^{(N-1)})^{0}\\
        (x^{(N)})^{d} & (x^{(N)})^{d-1} & \dots & (x^{(N)})^{1}& (x^{(N)})^{0}
    \end{bmatrix}
\label{eq8}
\tag{8}
\end{equation*}

After this we will combine this with our linear map (used in linear regression) $g^{(\mathbf{w})}=\mathbf{x}\rightarrow\mathbf{w}^T\mathbf{x}$:

\begin{equation*}
    h^{(\mathbf{w})}=g^{(\mathbf{w})}(\phi(\mathbf{x}))
\label{eq9}
\tag{9}
\end{equation*}

## 3.5 Behavior of GD

Recall step `3.3 Linear Regression` where the optimal predictor was solved by minimizing empirical error $\mathcal{E}$ (see ([2](#mjx-eqn-eq2))) and getting a closed form expression (equation 5). Now we will introduce another minimization method called gradient descent **(GD)**. 

Gradient descent amounts to iteratively updating a current guess $w^{(k)}$ for the optimal weight (which minimizes ([2](#mjx-eqn-eq2))) according to the rule:

\begin{equation*}
    \mathbf{w}^{(k+1)} = \mathbf{w}^{(k)} - \alpha {\nabla}_{\mathbf{w}}\mathcal{E} (\mathbf{w})
    \label{eq10}
    \tag{10}
\end{equation*}

Here $\alpha$ ($>0$) is a design parameter which is referred to as "step-size" or "learning-rate". The choice of the value of $\alpha$ is crucial for how quickly (and if at all) the iterates $w^{(k)}$ converge to the optimal weight vector $w_{opt}$. 

Note that the implementation of GD requires to compute the gradient ${\nabla}_{\mathbf{w}}\mathcal{E} (\mathbf{w})$.
For linear regression and squared error loss this gradient is obtained as 

\begin{equation*}
{\nabla}_{\mathbf{w}}\mathcal{E} (\mathbf{w}) =-\frac{2}{N}\sum^{N}_{i=1}x^{(i)}(y^{(i)} - {{\mathbf{w}}}^{T} x^{(i)}) = -\frac{2}{N} \mathbf{X}^{T}(\mathbf{y}-\mathbf{X} \mathbf{w})
\label{eq11}
\tag{11}
\end{equation*}

GD is started by initalizing weight vector $\mathbf{w}$ to 0 (i.e $\mathbf{w}^{(0)}=\begin{bmatrix}0,\dots,0\end{bmatrix}$). It's to be shown that for a suitably chosen learning rate, GD always converges to the optimal weight regardless of which inital guess $w^{(0)}$ has been used. However, the converge speed depends crucially on the precise choice of the learning rate. Especially __note__ that if $\alpha$ is too large, the iterations may not converge and if $\alpha$ is too small, the convergence is slow.

```python
# Calculate empirical error of the prediction
def empirical_risk(X, Y, w_opt):
    # Assign current weight vector according to iterative weight vector
    ### STUDENT TASK ###
    ## Compute empirical error by replacing '...' with your solution.
    ## Hints! Use self.X, self.Y and self.w_opt to get necessary matrices.
    ##        Check out numpy's dot(), mean(), power() and subtract() functions.
    # self.empirical_error = ... 
    empirical_error = np.mean(np.power(np.subtract(np.dot(X,w_opt), Y),2))
    return empirical_error

# Compute gradient
# return numpy.ndarray:
#      [ [w_1]
#          ....
#        [w_d]]
def gradient(X,Y,w):
    N = X.shape[0]
    ### STUDENT TASK ###
    ## Compute gradient by replacing '...' with your solution.
    ## Hint! Use self.X and self.Y to get necessary matrices.
    # return ...
    return np.dot(-2/N,np.dot(np.transpose(X),np.subtract(Y,np.dot(X,w))))

# Run GD for k steps
# a = alpha/learning rate
# k = iteration steps
# returns empirical error array with lenght k: [ Empirical_1, ..., Empirical_k ]
def gradient_descent(x, y, a, k):
    ### STUDENT TASK ###
    ## Hints! Same as in linearRegression()!
    # X = ...
    # Y = ...
    X = featureMatrix(x)
    Y = labelVector(y)
    # Initial weigth vector (all values 0)
    w = np.zeros((X.shape[1], Y.shape[1]))
    empirical_errors = []

    for i in range(k):
        #Calculate gradient
        grad = gradient(X,Y,w)

        ### STUDENT TASK ###
        ## Update weight vector by replacing '...' with your solution.
        # w = ...
        w = w - a * grad

        #Calculate Empirical Risk
        empirical = empirical_risk(X,Y,w)
        empirical_errors.append(empirical) 
    return empirical_errors
    
learning_rates = [0.001, 0.01, 0.02, 0.1, 0.2, 0.3, 0.9, 0.98]

# Compute gradient descent for different alpha values  
# and show learning curve(s) in the plot
for learning_rate in learning_rates:
    learning_curve=gradient_descent(x,y,learning_rate, 1000)   
    plt.plot(learning_curve,label=(r'$\alpha=$'+str(learning_rate)))  

plt.title(r'$\bf{Figure\ 6.}$Converge of GD')
plt.xlabel('Iterator k')
plt.ylabel('Empirical')
plt.legend()
plt.show()

learning_rates = [0.001, 0.01, 0.02, 0.1, 0.2, 0.3, 0.9, 0.98]

### STUDENT TASK ###
## Figure out the best value of alpha to get the fastest converge of GD
best_alpha = 0.9

# Compute gradient descent for different alpha values  
# and show learning curve(s) in the plot
for learning_rate in learning_rates:
    learning_curve=gradient_descent(x,y,learning_rate, 1000)   
    if learning_rate == best_alpha:
        plt.plot(learning_curve,label=(r'$\alpha=$'+str(learning_rate)), color="red")
    else:
        plt.plot(learning_curve,label=(r'$\alpha=$'+str(learning_rate)), color="blue")

plt.title(r'$\bf{Figure\ 7.}$Converge of GD')
plt.xlabel('Iterator k')
plt.ylabel('Empirical')
plt.legend()
plt.show()
```
