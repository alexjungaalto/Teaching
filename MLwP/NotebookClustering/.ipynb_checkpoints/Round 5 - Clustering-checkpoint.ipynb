##  1. Introduction
<a id="intro"></a>

On a high level, clustering is the task of dividing a (unlabeled) dataset $\mathbf{X} = \{ \mathbf{x}^{(i)} \}_{i=1}^{N}$, with $N$ data points $\mathbf{x}^{(i)} \in \mathbb{R}^{d}$, into a small number of groups or "clusters" $\mathcal{C}_{1},\ldots,\mathcal{C}_{k}$. Each cluster represents a subset of data points such that data points belonging to the same cluster are more similar to each other than to data points from another cluster. In contrast to regression and classification problems considered in earlier exercises, clustering methods do not require labeled data and can be applied to datasets whose data points $\mathbf{z}^{(i)}$ are characterized solely by its features $\mathbf{x}^{(i)}$. Therefore clustering methods are referred to as **unsupervised** machine learning methods.

There are two main flavours of clustering methods: 

* hard clustering methods  
* soft clustering methods


Hard clustering methods assign each data point to one and only one cluster. In contrast, soft-clustering methods assign each data point to several different clusters with varying degrees of belonging.

We will apply one popular method for hard clustering (k-means) and one popular method for soft clustering (which is based on a probabilistic Gaussian mixture model) to a real-life application. In particular, consider you are running a Cafe in Helsinki and you want to segment costumers in order to design a new marketing strategy for the upcoming summer. Such a costumer segmentation can be done efficiently using clustering methods.

## 3.1 Hard clustering with k-means
<a id="kmeans"></a>

A popular method for hard clustering is the k-means algorithm. K-means takes as input a list of data points $\mathbf{x}^{(1)},...,\mathbf{x}^{(N)} \in \mathbb{R}^{d}$ and groups them into $k$ non-overlapping clusters $\mathcal{C}_{¡},\ldots,\mathcal{C}_{k}$. Each cluster $\mathcal{C}_{c}$ is characterized by its cluster mean $\mathbf{m}_c \in \mathbb{R}^{d}$. As a hard-clustering method, k-means assigns each data point $\mathbf{x}^{(i)}$ to exactly one cluster whose index we denote as $y^{(i)} \in 1,...,k$. We can interpret the cluster assignments $y^{(i)}$ as (hidden) labels of the data points. However, we do not have access to the cluster assignment of any data point but rather have to estimate them using $k$-means. 


K-means can be summarized as follows:

* Choose initial cluster means $ \mathbf{m}_1,...,\mathbf{m}_k$

* Repeat until stopping condition is reached:  

    * Assign each datapoint to the cluster whose mean is nearest. 
    
    I.e. for all $i=1,...,N$, do  $$ y^{(i)} = \underset{c'}{\operatorname{argmin}} \|\mathbf{x}^{(i)} - \mathbf{m}^{(c')}\|^2 $$
    
    * Set the new cluster means of the clusters by calculating the average of the points in the cluster. 
    
    I.e. for all $c=1,...,k$, set $$ \mathbf{m}^{(c)} = \frac{1}{\mid\{i: y^{(i)}= c\}\mid}{\sum_{i: y^{(i)}= c}\mathbf{x}^{(i)}}     \label{mean}
    \tag{1}$$ 
    where $\{i: y^{(i)}= c\}$ represents the set of datapoints belonging to cluster c and $\mid\{i: y^{(i)}= c\}\mid$ the number of datapoints belonging to cluster c.  
    
    Note: Do this only for the clusters with at least one member, those with no members keep the old mean. 


Thus, k-means consists of 4 simple steps:

* Step 1 - Initialize cluster means.
* Step 2 - Assign datapoints to the nearest cluster means.
* Step 3 - Set new cluster means.
* Step 4 - If not finished go to step 2.

The working of $k$-means is best understood by walking through an example. To this end, we apply $k$-means to the cafe costumer data. 

**Step 1: Initialize cluster means**

We first initialize the cluster means, also called centroids. We do this by randomly choosing points between the minimum and maximum values of the data.

#### Tasks
- Implement a Python function `select_centroids()` which takes as inputs the datamatrix $\mathbf{X}=\big(\mathbf{x}^{(1)},\ldots,\mathbf{x}^{(N)}\big) \in \mathbb{R}^{d \times N}$ 
and the number $k$ of clusters and returns the initial centroids. 

- Hint: You can use `np.random.uniform()` for selection random numbers from some interval of values. 

```python
def select_centroids(data, k):
    ### STUDENT TASK ###
    
    centroids = np.zeros((k, data.shape[1]))
    
    for i in range(data.shape[1]):
        centroids[:,i] = np.random.uniform(np.min(data[:,i]), 
                                           np.max(data[:,i]), 
                                           size = (k))  
    
    return centroids
```

** Step 2: Cluster assignment update: Assign each datapoint to the cluster of nearest cluster mean **

In this step your task is to assign each data point $\mathbf{x}^{(i)}$ to the cluster $c$ with the nearest cluster mean $\mathbf{m}^{(c)}$. This distance is measured by the Euclidean distance $\| \mathbf{x}^{(i)} - \mathbf{m}^{(c)} \|$. This will result in a vector of length $N$ whose entries are the cluster assignments $y^{(i)}$ for each point.

#### Tasks
- Implement a Python function `assign_points()` which takes as inputs the data and centroids (cluster means) and returns a vector, _clusters_, whose entries are the cluster assignments $y^{(i)}$ for each point. 
    - You can use `np.linalg.norm()` to get the Euclidean distance. 
    - You can use `np.argmin()` to get the indices of the minimum values along an axis in an array.
    - Note: Remember that indexing in Python starts from 0, hence it might be easier to number the clusters from 0...k-1, but it is up to you to decide :)  

```python
def assign_points(data, centroids): 
    ### STUDENT TASK ###
    
    clusters = np.zeros(data.shape[0])
    
    for i in range(data.shape[0]):
        clusters[i] = np.argmin([np.linalg.norm(data[i] - centroids[j]) for j in range(centroids.shape[0])])
    
    return clusters
```

** Step 3: update cluster means **

We now want to update the cluster means $\mathbf{m}^{(c)}$ to be "in the middle" of the data points that were assigned to cluster c during the previous step. This "middle" is defined as the average (mean) of the data points belonging to the cluster, see equation (\ref{mean}). 

#### Tasks
- Implement the Python function `ḿove_centroids()` which takes as input the data, old centroids and cluster assignments. It should return the new centroids. 
    - You can use `np.mean()` to compute the mean of all points belong to one cluster. 
    - You need to keep in mind that there might be a situation when none of the points were assigned to a cluster. If this is the case, you should assign the old centroid as new centroid.

```python
def move_centroids(old_centroids, data, clusters):
    ### STUDENT TASK ###
    new_centroids = np.zeros(old_centroids.shape)
    for i in range(len(old_centroids)):
        cluster_points = np.array([data[j] for j in range(len(data)) if clusters[j] == i])
        if len(cluster_points) != 0:
            new_centroids[i] = np.mean(cluster_points, axis = 0)
        else:
            new_centroids[i] = old_centroids[i]
    return new_centroids
```

### 3.2. Handling local minima in k-means
<a id="local"></a>

As introduced in the course book, the k-means algorithm aims at minimizing the __empirical risk__: 

$$\mathcal{E}  ( \{\mathbf{m}_{c}\}_{c=1}^{k},\{y^{(i)}\}_{i=1}^{N} \mid \{\mathbf{x}^{(i)}\}_{i=1}^{N} )
=\frac{1}{N} \sum_{i=1}^{N} {\left\|\mathbf{x}^{(i)}-\mathbf{m}_{y^{(i)}}\right\|^2}
$$

Since the empirical risk is a highly non-convex function of the cluster means and assignments, the k-means method will sometimes get trapped in a local minimum. This especially happens when increasing the number of degrees. 

It is therefore useful to run k-means several times with different initializations for the cluster means and choose the cluster assignment that yields the smallest empirical risk. 

Tasks
- Implement a Python function `empirical_risk()` which takes as input the data, clusters and centroids and returns the empirical risk. 
- Then run the cell below, which iterates k-means 50 times and plots the best and worst cluster division. We use k=3, since this is a more difficult optimization problem than k=2 and hence gets more often stuck in local minima. 
<!--- Implement a Python function `new_k_means()` which takes as input the data and the number of clusters, k. It should return the cluster collection with the minimum empirical risk. 
    - Use 50 iterations of k-means and choose the iteration with the minimum empirical risk. 
    - You can reuse the function `k_means()` from the previous exercise. -->

```python
def empirical_risk(data, clusters, centroids):  
    ### STUDENT TASK ###
    
    risk = 0
    for i in range(len(data)):
        risk += np.linalg.norm(data[i,:] - centroids[int(clusters[i]),:])**2
    risk=risk/len(data)
   
    return risk
```

```python
def new_k_means(data, k,plot=True):  
    
    # initializing the array where we collect all cluster assignments  
    cluster_collection = np.zeros((50, data.shape[0]))
    # initializing the array where we collect all risk values 
    risk_collection = np.zeros(50)
    
    for i in range(50):
        centroids,clusters=k_means(data,k,plot=False)
        risk_collection[i] = empirical_risk(data, clusters, centroids)
        cluster_collection[i,:] = clusters
        #plotting(data,clusters=clusters)
        
    #find the best cluster assignment and print the lowest found empirical risk
    min_ind = np.argmin(risk_collection)
    max_ind=np.argmax(risk_collection)
    if plot==True:
        print("Cluster division with lowest empirical risk")
        plotting(data,clusters=cluster_collection[min_ind,:])
        print("Cluster division with highest empirical risk")
        plotting(data,clusters=cluster_collection[max_ind,:])
    
        print('min empirical risk is ', np.min(risk_collection))
    return cluster_collection[min_ind,:],risk_collection
    
best_cluster,risk=new_k_means(data,3)
```

## 3.2 Soft clustering with Gaussian Mixture Models (GMM) 
<a id="GMM"></a>

The information provided by $k$ means is rather coarsei grained: even if two data points belong to the cluster, their location relative to the other clusters might be very different. Let us now consider soft-clustering methods which provdide a more fine-grained information about the cluster structure of a data set. 

Considering again the costumer segmentation for a Cafe business, we might like to have some measure for the extend (or degree) by which a customer belongs to various groups. This is a soft-clustering problem where we associate each data point $\mathbf{x}^{(i)}$(which represents a particular customer) with a membership-vector $\mathbf{y}^{(i)}= (y^{(i)}_1,...,y^{(i)}_k) \in [0,1]^k$ whose entry $y^{(i)}_c$ is the degree by which we assign $\mathbf{x}^{(i)}$ to cluster $\mathcal{C}_c$. 

A principled approach to obtaining a soft-clustering method is based on interpreting the data points $\mathbf{x}^{(i)}$ as realizations of a random vector $\mathbf{x}$ with some underlying probablitiy distribution $p(\mathbf{x})$. In particular, we can represent a cluster $\mathcal{C}_{c}$ by a Gaussian distribution with mean vector $\mathbf{m}^{(c)}$ and covariance matrix $\mathbf{C}^{(c)}$. The probability density function of such a Gaussian distribution is denoted 

\begin{equation}
\mathcal{N}(\mathbf{x} ; \mathbf{m}^{(c)}, \mathbf{C}^{(c)}) = \frac{1}{\sqrt{{\rm det} \big(2 \pi \mathbf{C}^{(c)}\big)}} {\rm exp } \bigg( - \big(\mathbf{x} - \mathbf{m}^{(c)} \big)^{T} \big(\mathbf{C}^{(c)}\big)^{-1}  \big(\mathbf{x} - \mathbf{m}^{(c)} \big)\bigg).
\end{equation}

The overall probability distribution is then obtained as a **Gaussian mixture** 

\begin{equation}
p(\mathbf{x}) = p_{1} \mathcal{N}(\mathbf{x} ; \mathbf{m}^{(1)}, \mathbf{C}^{(1)}) + p_{2} \mathcal{N}(\mathbf{x} ; \mathbf{m}^{(2)}, \mathbf{C}^{(2)})
\end{equation}

with coefficients $p_{1},p_{2} \geq 0$ satisfying $p_{1} + p_{2} =1$. The coefficient $p_{1}$ is the probability that a data point is drawn from cluster $\mathcal{C}_{1}$ and, similarly, $p_{2}$ is the probability that a data point is drawn from cluster $\mathcal{C}_{2}$. Note that the distribution $p(\mathbf{x})$ depends (is parametrized by) the coefficients $p_{1},p_{2}$, the cluster means $\mathbf{m}^{(1)},\mathbf{m}^{(2)}$ 
and the covariance matrices $\mathbf{C}^{(1)},\mathbf{C}^{(2)}$. 

Under the probabilistic model $p(\mathbf{x})$, the degree $y_{c}^{(i)}$ of a data point belonging to cluster $\mathcal{C}_{c}$ can be defined as the probability that $\mathbf{x}^{(i)}$ is generated (drawn) from the Gaussian distribution associated with $\mathcal{C}_{c}$:  

$$\mathbf{y}^{(i)}_c = \frac{\mathcal{N}(\mathbf{x}^{(i)} ; \mathbf{m}^{(c)}, \mathbf{C}^{(c)})}{\sum_{c'=1}^k \mathcal{N}(\mathbf{x}^{(i)} ; \mathbf{m}^{(c')}, \mathbf{C}^{(c')})} $$

After determining the degrees of belonging, we can then update or guess (estimate) for the cluster probabilities $p_{1},p_{2}$, cluster means $\mathbf{m}^{(1)}, \mathbf{m}^{(2)}$ and covariance matrices $\mathbf{C}^{(1)},\mathbf{C}^{(2)}$.

In summary this algorithm consists of 4 steps (just as k-means):

* Step 1 - Initailize the cluster parameters. This are the means and covariances for every cluster. 
* Step 2 - Assign a probability for each datapoint belonging to each cluster. 
* Step 3 - Set the new cluster parameters, i.e. the means and covariances. 
* Step 4 - Put steps 2 and 3 in a loop

So lets go throught these steps one by one and compose the algorithm. 

** Step 1: Initialize the cluster parameters**

We first choose an initial mean and covariance matrix for each cluster. To this end, we initialize the mean in the same way as in $k$-means. As initialization for the covariance matrix, we take the identity matrix. 

Tasks
- Implement a Python function `initialize_parameters()` which takes as input the data and the number $k$ of clusters  and returns the initial means and covariance matrices for all clusters. 
- Hint: You can reuse the code of `select_centroids()` of the k-mean algorithm to get the initial means


```python
def initialize_parameters(data, k):
    ### STUDENT TASK ###
    
    means = np.zeros((k, data.shape[1]))
    for i in range(data.shape[1]):
        means[:,i] = np.random.uniform(np.min(data[:,i]), 
                                           np.max(data[:,i]), 
                                           size = (k)) 
    covariances=np.zeros((k,data.shape[1],data.shape[1]))
    
    for f in range(k):
        covariances[f]=np.identity(data.shape[1])
    return means, covariances
```

```python
def plot_GMM_default(data,means,covariances,k,cluster_vectors=None):
    if cluster_vectors is None:
        plt.scatter(data[:,0], data[:,1], s=13,alpha=0.5)
    else:
        clusters = np.argmax(cluster_vectors,axis=0)
       
        #plt.colormap
        plt.scatter(data[:,0], data[:,1], c=clusters, s=13,alpha=0.5)
    
    color=['r','g','b']
    #Visualization of results
    x_plot = np.linspace(-9,3, 100)
    y_plot = np.linspace(-3,7, 100)

    for i in range(k):
        x_mesh, y_mesh = np.meshgrid(x_plot, y_plot)
        z= plt.mlab.bivariate_normal(x_mesh, y_mesh, np.sqrt(covariances[i,0, 0]), \
                                    np.sqrt(covariances[i,1, 1]), means[i,0], means[i,1],covariances[i,0,1])
        plt.contour(x_mesh , y_mesh , z,4,colors=color[i])
        plt.scatter( [means[i,0]], [means[i,1]], marker='x',c='red')

    plt.title("Soft clustering with GMM")
    plt.xlabel("feature x_1: customers' age")
    plt.ylabel("feature x_2: money spent during visit")
    plt.show()
```

```python
means,covariances=initialize_parameters(data,3)
plot_GMM_default(data,means,covariances,3)
```

** Step 2: Assign a probability for each datapoint belonging to each cluster ** 

We now want to calculate for each datapoint the probability it belongs to each cluster. The probability of belonging to a cluster, is also called the degree of beloninging. We can calculate this by the following formula:

$$y^{(i)}_c = \frac{\mathcal{N}(\mathbf{x}^{(i)} ; \mathbf{m}_c, \mathbf{C}_c)}{\sum_{j=1}^k \mathcal{N}(\mathbf{x}^{(i)} ; \mathbf{m}_j, \mathbf{C}_j)} $$

$\mathcal{N}(\mathbf{x}^{(i)} ; \mathbf{m}_c, \mathbf{C}_c)$ is called the probability density function (or pdf for short). 

Tasks
- Implement a Python function `update_degrees_of_belonging()` which takes as inputs the data, means, covariances and the number of clusters (k) and returns the cluster vectors, $\mathbf{y}$. 
    - In order to achieve this, you should implement the formula given above. 
    - You can calculate the pdf by using `multivariate_normal.pdf()` from the scipy.stats library

```python
from scipy.stats import multivariate_normal # Multivariate normal random variable


def update_degrees_of_belonging(data, means, covariances,k): 
    ### STUDENT TASK ###
    cluster_vectors=np.zeros((k,data.shape[0]))   
    for i in range(data.shape[0]):
        belonging_all=np.sum([multivariate_normal.pdf(data[i],means[f],covariances[f]) for f in range(k)])
        for t in range(k):
            cluster_vectors[t,i]=multivariate_normal.pdf(data[i],means[t],covariances[t])/belonging_all
    
    return cluster_vectors
```

```python
cluster_vectors=update_degrees_of_belonging(data,means,covariances,3)
plot_GMM_default(data,means,covariances,3,cluster_vectors)
```

** Step 3: Set the new cluster parameters **

Now we want to update the cluster parameters. In contrast to $k$-means, where we only updated the clusters means $\mathbf{m}^{(1)},\mathbf{m}^{(2)}$, here we also have to update the covariances. Since we use degrees of belonging the formula for updating the means is slightly different. See both updating formulas below. 

$$\mathbf{m}^{(c)} = \frac{1}{N_c} \sum_{i=1}^N y_c^{(i)}\mathbf{x}^{(i)} $$
$$\mathbf{C}^{(c)} = \frac{1}{N_c} \sum_{i=1}^N y_c^{(i)}(\mathbf{x}^{(i)} - \mathbf{m}_c)(\mathbf{x}^{(i)} - \mathbf{m}_c)^T $$
, where $M_c = \sum_{i=1}^N y_c^{(i)}$

Tasks
- Implement a Python function `update_GMM_pars()` which takes as inputs the data, the cluster vectors and the number of clusters. It should return the updated mean and covariance matrix for every cluster. 

```python
# Explain where the update equation came from?
```

```python
def update_GMM_pars(data, cluster_vectors,k): 
    means_new = np.zeros((k, data.shape[1]))
    covariances_new=np.zeros((k,data.shape[1],data.shape[1]))
    for i in range(k):
        sum_k=np.sum(cluster_vectors[i])
        means_new[i]=data.T@cluster_vectors[i]/sum_k
        for l in range(data.shape[0]):
            covariances_new[i]+=np.outer( data[l] - means_new[i], data[l] - means_new[i]) * cluster_vectors[i,l]
        covariances_new[i]=covariances_new[i]/sum_k
    return means_new,covariances_new
```

```python
means,covariances=update_GMM_pars(data, cluster_vectors,3)
plot_GMM_default(data,means,covariances,3,cluster_vectors)
```

** Step 4: Put steps 2 and 3 in a loop **

The last step is to put all the pieces together by initializing the parameters and then putting step 2 and 3 inside a loop. This is very similar to the process in k-means. 

Tasks
- Implement a Python function `GMM_clustering()` which takes as inputs the data, the number of clusters and the number of iterations. It should return the final means and covariance matrices. 
    - It is the same as step 4 in k-means, only now we use the new functions we created for soft-clustering. 

```python
def GMM_clustering(data,k,num_iters):
    ### STUDENT TASK ###
    # Step 1: 
    means,covariances = initialize_parameters(data,k)

    for i in range(num_iters):
        # Step 2:
        cluster_vectors = update_degrees_of_belonging(data, means, covariances,k)
        # Step 3:
        means,covariances=update_GMM_pars(data, cluster_vectors,k)
    
    return means, covariances, cluster_vectors
```

Now run the code to calculate the final means and covariance matrices and plot the result! The red circles indicate the covariance matrices of each cluster. 

```python
def plot_GMM(data,k,num_iter):
    means,covariances,cluster_vectors=GMM_clustering(data,k,num_iter)
    print("The means are",means)
    print("The covariance matrices are",covariances)
    clusters = np.argmax(cluster_vectors,axis=0)
    plt.figure(figsize=(6,6))
    #plt.colormaps
    color=['r','g','b']
    #plt.colormap
    plt.scatter(data[:,0], data[:,1], c=clusters, s=13,alpha=0.5)
    

    #Visualization of results
    x_plot = np.linspace(-9,3, 100)
    y_plot = np.linspace(-3,7, 100)

    for i in range(k):
        x_mesh, y_mesh = np.meshgrid(x_plot, y_plot)
        z= plt.mlab.bivariate_normal(x_mesh, y_mesh, np.sqrt(covariances[i,0, 0]), \
                                    np.sqrt(covariances[i,1, 1]), means[i,0], means[i,1],sigmaxy=covariances[i,0,1])
        plt.contour(x_mesh , y_mesh , z,4,colors=color[i])
        plt.scatter( [means[i,0]], [means[i,1]], marker='x',c='red')

    plt.title("Soft clustering with GMM")
    plt.xlabel("feature x_1: customers' age")
    plt.ylabel("feature x_2: money spent during visit")
    plt.show()
    
plot_GMM(data,3,50)
```
